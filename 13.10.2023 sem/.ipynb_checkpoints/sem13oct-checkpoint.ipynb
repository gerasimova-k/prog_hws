```python
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np
import nltk
from nltk import tokenize
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.probability import FreqDist
```

```python
def calculate_word_frequencies(raw):
    punctuations = ".?!,:;’-–—”()[]{}<>“/…*&#~\@^|«»"
    txt_no_punc = ""
    text = raw.replace('\n', ' ')
    for char in text:
        if char not in punctuations:
            txt_no_punc += char

    all_words = word_tokenize(txt_no_punc.lower())
    frequency = Counter(all_words)
    number_of_words = len(all_words)
    frequency_dict = dict(frequency)
    sorted_dict = {k: v for k, v in sorted(frequency_dict.items(), key=lambda item: item[1], reverse=True)}
    lst_sorted_dict = [(k, v) for (k, v) in sorted_dict.items()]
    top_words = list(sorted_dict.keys())[:10]
    x = top_words
    y = [sorted_dict[word] for word in list(sorted_dict.keys())[:10]]
    labels = top_words
    plt.bar(x, y)
    plt.title('Top 10 Word Frequencies')
    # You can specify a rotation for the tick labels in degrees or with keywords.
    plt.xticks(x, labels, rotation='vertical')
    # Pad margins so that markers don't get clipped by the axes
    plt.margins(0.2)
    # Tweak spacing to prevent clipping of tick-labels
    plt.subplots_adjust(bottom=0.15)
    plt.show()
```

```python
years = list(range(2001, 2021))
rand_data = [random.randint(0, 100) for _ in range(len(years))]

x1 = []
x2 = []
x3 = []
x4 = []

y1 = []
y2 = []
y3 = []
y4 = []

for year, value in zip(years, rand_data):
    if 2001 <= year <= 2005:
        y1.append(value)
        x1.append(year)
    if 2006 <= year <= 2010:
        y2.append(value)
        x2.append(year)
    if 2011 <= year <= 2015:
        y3.append(value)
        x3.append(year)
    if 2016 <= year <= 2020:
        y4.append(value)
        x4.append(year)

def make_plot(axs):
    box = dict(facecolor='yellow', pad=5, alpha=0.2)

    ax1 = axs[0, 0]
    ax1.plot(x1, y1)
    ax1.set_ylabel('2001-2005', bbox=box)
    ax1.set_ylim(0, 100)
    ax1.set_xlim(2001, 2005)

    ax2 = axs[0, 1]
    ax2.set_ylabel('2006-2010', bbox=box)
    ax2.plot(x2, y2)
    ax2.set_ylim(0, 100)
    ax2.set_xlim(2006, 2010)

    ax3 = axs[1, 0]
    ax3.plot(x3, y3)
    ax3.set_ylabel('2011-2015', bbox=box)
    ax3.set_ylim(0, 100)
    ax3.set_xlim(2011, 2015)

    ax4 = axs[1, 1]
    ax4.set_ylabel('2016-2020', bbox=box)
    ax4.plot(x4, y4)
    ax4.set_ylim(0, 100)
    ax4.set_xlim(2016, 2020)
    

fig, axs = plt.subplots(2, 2)
fig.subplots_adjust(left=0.2, wspace=0.6)
make_plot(axs)


plt.show()

```

```python
def plot_histogram(x, n):
    my_parameters = range(n)
    my_values = [random.randint(0, x) for _ in range(n)]

    
    sns.set_style("dark")
    
    sns.barplot(my_values)
    
 
    plt.xlabel("Parameter")
    plt.ylabel("Value")
    
    plt.show()
```

```python
plot_histogram(10, 12)
```

```python
interact(plot_histogram, x=IntSlider(min=1, max=30, step=1, value=1), n=list(range(1, 11)))
```
