```python
from nltk.util import ngrams
import collections

punctuations = ".?!,:;’-–—”()[]{}<>“/…*&#~\@^|«»"
txt_no_punc = ""

with open('Янь Мо. Устал рождаться и умирать.txt', 'r', encoding='utf-8') as file:
    read_file = file.read().lower().replace('\n', ' ')
    for char in read_file:
        if char not in punctuations:
            txt_no_punc += char

    text = nltk.Text(nltk.word_tokenize(txt_no_punc))

    all_n_grams = list(ngrams(text, 4))

    frequency = collections.Counter(all_n_grams)
    frequency_dict = dict(frequency)
    sorted_dict = {k: v for k, v in sorted(frequency_dict.items(), key=lambda item: item[1], reverse=True)}
    lst_sorted_dict = [(k, v) for (k, v) in sorted_dict.items()]

    print(lst_sorted_dict[0:5])
```

```python
text = '''“那当然。如果你听话的话，我再给你画一根绳子，白天可以栓住它。再加上一根扦杆。”

我的建议看来有点使小王子反感。

“栓住它，多么奇怪的主意。”

“如果你不栓住它，它就到处跑，那么它会跑丢的。”

我的这位朋友又笑出了声：

“你想要它跑到哪里去呀？”

“不管什么地方。它一直往前跑……”

这时，小王子郑重其事地说：

“这没有什么关系，我那里很小很小。”

接着，他略带伤感地又补充了一句：

“一直朝前走，也不会走出多远……”'''
nlp = spacy.load("zh_core_web_sm")
doc = nlp(text)
for ent in doc:
    print(ent.text,ent.tag_,ent.morph)
```

```python
import collections
```

```python
raw = '''当我还只有六岁的时候，在一本描写原始森林的名叫《真实的故事》的书中，看到了一幅精彩的插画，画的是一条蟒蛇正在吞食一只大野兽。页头上就是那幅画的摹本。

这本书中写道：“这些蟒蛇把它们的猎获物不加咀嚼地囫囵吞下，尔后就不能再动弹了；它们就在长长的六个月的睡眠中消化这些食物。”

当时，我对丛林中的奇遇想得很多，于是，我也用彩色铅笔画出了我的第一幅图画。我的第一号作品。它是这样的：

我把我的这幅杰作拿给大人看，我问他们我的画是不是叫他们害怕。

他们回答我说：“一顶帽子有什么可怕的？”'''

punctuations = "，。《》：“”；？"
txt_no_punc = ""
text = raw.replace('\n', ' ')
for char in text:
    if char not in punctuations:
        txt_no_punc += char

nlp = spacy_udpipe.load('zh')
doc = nlp(txt_no_punc)

all_lemmas = []
for token in doc:
    all_lemmas.append(token.lemma_)

frequency = collections.Counter(all_lemmas)
frequency_dict = dict(frequency)
sorted_dict = {k: v for k, v in sorted(frequency_dict.items(), key=lambda item: item[1], reverse=True)}
lst_sorted_dict = [(k, v) for (k, v) in sorted_dict.items()]

print(lst_sorted_dict[0:5])
```
