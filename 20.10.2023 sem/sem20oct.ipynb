```python
russia_map = folium.Map(
    location = [64.6863136, 97.7453061],    
    zoom_start = 4
)

formatter = "function(num) {return L.Util.formatNum(num, 5);};" # это проще копировать
mouse_position = MousePosition(
    position='topright', # где будет находиться окошко с координатами
    separator=' Long: ', # долгота
    empty_string='NaN', # что будет показываться, когда координат нет
    lng_first=False,
    num_digits=20, # сколько цифр влезет
    prefix='Lat:', # широта
    lat_formatter=formatter,
    lng_formatter=formatter,
)

russia_map.add_child(mouse_position)
russia_map
    
```

```python
try:  
    original = Image.open('statue.jpg')  
except FileNotFoundError:  
    print("Файл не найден")
```

```python
from PIL import ImageFilter
```

```python
# размываем изображение
blurred = original.filter(ImageFilter.BLUR)
# открываем оригинал и размытое изображение
original.show() # покажет картинку-оригинал
blurred.show() # покажет, что получилось
# сохраняем изображение
blurred.save("blurred.png")
```

```python
contour = original.filter(ImageFilter.CONTOUR)
contour.show() 
contour.save("contour.png")
```

```python
detail = original.filter(ImageFilter.DETAIL)
detail.show() 
detail.save("detail.png")
```

```python
edge_enhance = original.filter(ImageFilter.EDGE_ENHANCE)
edge_enhance.show() 
edge_enhance.save("edge_enhance.png")
```

```python
emboss = original.filter(ImageFilter.EMBOSS)
emboss.show() 
emboss.save("emboss.png")
```

```python
# Помимо этого, можно использовать еще такие фильтры:

    BLUR
    CONTOUR
    DETAIL
    EDGE_ENHANCE
    EDGE_ENHANCE_MORE
    EMBOSS
    FIND_EDGES
    SMOOTH
    SMOOTH_MORE
    SHARPEN

```

```python
converted_img = original.transpose(Image.FLIP_LEFT_RIGHT)
converted_img.show()
converted_img.save("converted_left_right.png")
```

```python
import json 

trochaic = json.load(open('trochaic.geojson'))
lambic = json.load(open('lambic.geojson'))
no_rhythmic_stress = json.load(open('no_rhythmic_stress.geojson'))
dual = json.load(open('dual_both_trochaic_and_iambic.geojson'))
undetermined = json.load(open('undetermined.geojson'))
m = folium.Map(location=[50, 90], zoom_start=2)


folium.GeoJson(
    trochaic,
    zoom_on_click=True,
    marker=folium.Marker(icon=folium.Icon(color="pink",icon="heart", prefix='fa')),
).add_to(m)

folium.GeoJson(
    lambic,
    zoom_on_click=True,
    marker=folium.Marker(icon=folium.Icon(color="green",icon="music", prefix='fa')),
).add_to(m)

folium.GeoJson(
    no_rhythmic_stress,
    zoom_on_click=True,
    marker=folium.Marker(icon=folium.Icon(color="blue",icon="star", prefix='fa')),
).add_to(m)

folium.GeoJson(
    dual,
    zoom_on_click=True,
    marker=folium.Marker(icon=folium.Icon(color="red",icon="headphones", prefix='fa')),
).add_to(m)

folium.GeoJson(
    undetermined,
    zoom_on_click=True,
    marker=folium.Marker(icon=folium.Icon(color="purple",icon="gift", prefix='fa')),
).add_to(m)

m
```
